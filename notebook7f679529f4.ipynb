```python
import numpy as np

class SimpleNeuralNetwork:
    def __init__(self):
        # تأكد أن المعاملات التي تحتاجها في الكود موجودة في هذا المكان
        # مثال: حجم المدخلات (input_size)، حجم الطبقة المخفية (hidden_size)، وحجم الطبقة الناتجة (output_size)
        self.input_size = 2
        self.hidden_size = 2
        self.output_size = 1
        
        # تهيئة الأوزان العشوائية
        self.weights_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.weights_hidden_output = np.random.randn(self.hidden_size, self.output_size)
        
        # تهيئة الانحرافات (biases)
        self.bias_hidden = np.random.randn(1, self.hidden_size)
        self.bias_output = np.random.randn(1, self.output_size)
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def feedforward(self, X):
        # المرور الأمامي (التنبؤ)
        self.hidden_layer_activation = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        self.hidden_layer_output = self.sigmoid(self.hidden_layer_activation)
        self.output_layer_activation = np.dot(self.hidden_layer_output, self.weights_hidden_output) + self.bias_output
        output = self.sigmoid(self.output_layer_activation)
        return output

# استخدام الشبكة العصبية
nn = SimpleNeuralNetwork()

# مثال للمدخلات
X = np.array([[0, 1], [1, 1], [1, 0], [0, 0]])

# تنفيذ عملية التنبؤ
predictions = nn.feedforward(X)
print(predictions)
```
